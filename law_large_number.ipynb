# Law of large numbers

```python
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as plt
```

```python
N = 1000
mu = 2
sigma = 1
alpha = 0.02
```

$\mu_k^{(1)} = \frac{1}{k}\sum x_k$

$\mu_k^{(2)} = \mu_{k-1}^2+\frac{1}{k}(x_k-\mu_{k-1}^{(2)})$

$\mu_k^{(3)} = \mu_{k-1}^{(3)}+\alpha(x_k-\mu_{k-1}^{(3)})$

```python
x = rd.rand(N,1) * sigma + mu
# x = rd.randn(N,1) * sigma + mu

plt.plot(x, np.zeros_like(x)+0.1*rd.randn(N,1),'.')
plt.axis('square')
plt.axis([np.min(x)-1, np.max(x)+1, -1, 1])
plt.grid(True)
```

```python
mu1 = np.cumsum(x) / (np.arange(N) + 1)
mu2 = [x[0]]
mu3 = [x[0]]
for k in range(1,N):
    mu2.append(mu2[-1] + 1/k * (x[k] - mu2[-1]))
    mu3.append(mu3[-1] + alpha * (x[k] - mu3[-1]))
mu2 = np.array(mu2)
mu3 = np.array(mu3)
```

```python
n = np.arange(N)
plt.plot(n, x, linewidth=0.5)
plt.plot(n, mu1)
plt.plot(n, mu2)
plt.plot(n, mu3)
plt.grid(True)
```
